```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
j_df = pd.read_excel('Annual+DAF+Report+2025+Dataset.xlsx', sheet_name='data')


j_df_conf = j_df.copy()
```

```python
j_df_conf.loc[j_df_conf['source_record'] == "IRS 990", 'source_record'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_record'] == "NPT Historical Data", 'source_record'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_record'] == "Paper Submission", 'source_record'] = "Confirmed"

j_df_conf.loc[j_df_conf['source_number_of_accounts'] == "IRS 990", 'source_number_of_accounts'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_number_of_accounts'] == "NPT Historical Data", 'source_number_of_accounts'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_number_of_accounts'] == "Paper Submission", 'source_number_of_accounts'] = "Confirmed"

j_df_conf.loc[j_df_conf['source_total_daf_contributions'] == "IRS 990", 'source_total_daf_contributions'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_total_daf_contributions'] == "NPT Historical Data", 'source_total_daf_contributions'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_total_daf_contributions'] == "Paper Submission", 'source_total_daf_contributions'] = "Confirmed"

j_df_conf.loc[j_df_conf['source_total_daf_grants'] == "IRS 990", 'source_total_daf_grants'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_total_daf_grants'] == "NPT Historical Data", 'source_total_daf_grants'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_total_daf_grants'] == "Paper Submission", 'source_total_daf_grants'] = "Confirmed"

j_df_conf.loc[j_df_conf['source_eoy_daf_assets'] == "IRS 990", 'source_eoy_daf_assets'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_eoy_daf_assets'] == "NPT Historical Data", 'source_eoy_daf_assets'] = "Confirmed"
j_df_conf.loc[j_df_conf['source_eoy_daf_assets'] == "Paper Submission", 'source_eoy_daf_assets'] = "Confirmed"


j_df_conf = j_df_conf[(j_df_conf['source_record'] == 'Confirmed') & (j_df_conf['source_number_of_accounts'] == 'Confirmed') & (j_df_conf['source_total_daf_contributions'] == 'Confirmed') & (j_df_conf['source_total_daf_grants'] == 'Confirmed') & (j_df_conf['source_eoy_daf_assets'] == 'Confirmed')]
```

```python
filter_df19 = j_df_conf.loc[(j_df_conf["fiscal_year_end"] == 2019)]
filter_df20 = j_df_conf.loc[(j_df_conf["fiscal_year_end"] == 2020)]
filter_df21 = j_df_conf.loc[(j_df_conf["fiscal_year_end"] == 2021)]
filter_df22 = j_df_conf.loc[(j_df_conf["fiscal_year_end"] == 2022)]
filter_df23 = j_df_conf.loc[(j_df_conf["fiscal_year_end"] == 2023)]
filter_df24 = j_df_conf.loc[(j_df_conf["fiscal_year_end"] == 2024)]
```

```python
summary_data19 = filter_df19.describe()
summary_data20 = filter_df20.describe()
summary_data21 = filter_df21.describe()
summary_data22 = filter_df22.describe()
summary_data23 = filter_df23.describe()
summary_data24 = filter_df24.describe()

maxes = [summary_data19.loc["max","total_daf_contributions"],
          summary_data20.loc["max","total_daf_contributions"],
            summary_data21.loc["max","total_daf_contributions"],
              summary_data22.loc["max","total_daf_contributions"],
              summary_data23.loc["max","total_daf_contributions"], 
              summary_data24.loc["max","total_daf_contributions"]]
```

```python
l = [2019, 2020, 2021, 2022, 2023, 2024]
plt.bar(l, maxes)
plt.show()
```

```python
means = [summary_data19.loc["mean","total_daf_contributions"],
          summary_data20.loc["mean","total_daf_contributions"],
            summary_data21.loc["mean","total_daf_contributions"],
              summary_data22.loc["mean","total_daf_contributions"],
              summary_data23.loc["mean","total_daf_contributions"], 
              summary_data24.loc["mean","total_daf_contributions"]]

plt.bar(l, means)
plt.show()
```

```python
data = [
    filter_df19["total_daf_contributions"].dropna(),
    filter_df20["total_daf_contributions"].dropna(),
    filter_df21["total_daf_contributions"].dropna(),
    filter_df22["total_daf_contributions"].dropna(),
    filter_df23["total_daf_contributions"].dropna(),
    filter_df24["total_daf_contributions"].dropna(),
]

plt.figure(figsize=(10,6))
plt.boxplot(data)
plt.xticks([1,2,3,4,5,6], ["2019","2020","2021","2022","2023","2024"])
plt.show()
```

```python
mins = [summary_data19.loc["min","total_daf_contributions"],
          summary_data20.loc["min","total_daf_contributions"],
            summary_data21.loc["min","total_daf_contributions"],
              summary_data22.loc["min","total_daf_contributions"],
              summary_data23.loc["min","total_daf_contributions"], 
              summary_data24.loc["min","total_daf_contributions"]]

contributions_19_norm = (filter_df19["total_daf_contributions"] - mins[0])/(maxes[0] - mins[0])
contributions_20_norm = (filter_df20["total_daf_contributions"] - mins[1])/(maxes[1] - mins[1])
contributions_21_norm = (filter_df21["total_daf_contributions"] - mins[2])/(maxes[2] - mins[2])
contributions_22_norm = (filter_df22["total_daf_contributions"] - mins[3])/(maxes[3] - mins[3])
contributions_23_norm = (filter_df23["total_daf_contributions"] - mins[4])/(maxes[4] - mins[4])
contributions_24_norm = (filter_df24["total_daf_contributions"] - mins[5])/(maxes[5] - mins[5])
contributions_norm = [contributions_19_norm, contributions_20_norm, contributions_21_norm, contributions_22_norm, contributions_23_norm, contributions_24_norm]
norm_means = np.array([np.mean(contributions_19_norm), np.mean(contributions_20_norm), np.mean(contributions_21_norm), np.mean(contributions_22_norm), np.mean(contributions_23_norm), np.mean(contributions_24_norm)])
norm_stds = np.array([np.std(contributions_norm[0]), np.std(contributions_norm[1]), np.std(contributions_norm[2]), np.std(contributions_norm[3]), np.std(contributions_norm[4]), np.std(contributions_norm[5])])
```
