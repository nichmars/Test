```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
file = 'Filtered Data with Subtypes.xlsx'
df = pd.read_excel(file)
```

```python
#addition to drop the entries with N/A in the number of accounts
accounts_as_num = pd.to_numeric(df["Accounts"], errors="coerce")

print(accounts_as_num.isna().sum())

accounts_as_num = accounts_as_num.dropna()
print(accounts_as_num.isna().sum())

df["Accounts"] = accounts_as_num

contributions_as_num = pd.to_numeric(df["Contributions"], errors="coerce")

print(contributions_as_num.isna().sum())

contributions_as_num = contributions_as_num.dropna()
print(contributions_as_num.isna().sum())

df["Contributions"] = contributions_as_num

grants_as_num = pd.to_numeric(df["Grants"], errors="coerce")

print(grants_as_num.isna().sum())

grants_as_num = grants_as_num.dropna()
print(grants_as_num.isna().sum())

df["Grants"] = grants_as_num

assets_as_num = pd.to_numeric(df["Assets"], errors="coerce")

print(assets_as_num.isna().sum())

assets_as_num = assets_as_num.dropna()
print(assets_as_num.isna().sum())

df["Assets"] = assets_as_num
```

```python
df = df.drop_duplicates(subset=["ein", "Year"], keep="first")
#df = df.drop(columns = ["Source"])
df = df.dropna()
df.head(20)
```

```python
df.to_excel("Filtered Data with Subtypes Again.xlsx", index=False)
```
