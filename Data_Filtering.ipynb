```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression
from google.colab import drive
import matplotlib.pyplot as plt
drive.mount('/content/drive')
```

```python
file = '/content/drive/MyDrive/MTH498 Project/Combined Datasets.xlsx'
df = pd.read_excel(file)
```

```python
#addition to drop the entries with N/A in the number of accounts
df = df.dropna(subset=['Accounts'])
df = df[df['Accounts'] != 'No Data']
df = df[df['Accounts'] != 'No data']
df = df[df['Accounts'] != 'no data']
df = df[df['Accounts'] != 'Closed']
df = df[df['Accounts'] != 'closed']
df = df[df['Accounts'] != 'EZ']
df = df[df['Accounts'] != 'PF']
df = df[df['Accounts'] != 'Blank']
df = df[df['Accounts'] != 'blank']
df = df[df['Accounts'] != 0]
```

```python
df = df[df['Contributions'] != "blank"]
df = df[df['Contributions'] != "no data"]
df = df[df['Grants'] != "blank"]
df = df[df['Grants'] != "no data"]
df = df[df['Assets'] != "blank"]
df = df[df['Assets'] != "no data"]
```
