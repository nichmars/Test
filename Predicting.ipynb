```python
import pandas as pd
```

```python
df2=pd.read_excel('historical DAF sponsor data.xlsx' )
```

```python
CoFound = df2[df2['Type'] == 'Community Foundation']
CoFound_2024=CoFound['2024 Contributions'].dropna(axis=0)
CoFound_2024=[x for x in CoFound_2024 if type(x) is int]
tot_CoFound_2024 = sum(CoFound_2024)
print(tot_CoFound_2024)
CoFound_accounts=CoFound['2024 Funds'].dropna(axis=0)
CoFound_accounts=[x for x in CoFound_accounts if type(x) is int]
tot_CoFound_accounts=sum(CoFound_accounts)
print(tot_CoFound_accounts)
CoFound_predict=tot_CoFound_2024/tot_CoFound_accounts
print(CoFound_predict)
```
