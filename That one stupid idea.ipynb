```python
import pandas as pd
```

```python
df = pd.read_excel("Filtered Data with Subtypes.xlsx")
```

```python
df2 = df.copy()
df2finalrow = df2.tail(1)
df2 = df2.iloc[:-1]
df2 = pd.concat([df2finalrow, df2], ignore_index=True)
```

```python
sidebyside = pd.concat([df, df2], ignore_index=True, axis=1)
sidebyside.head()
```

```python
sidebyside.dtypes
sidebyside[0][3]
```

```python
sidebyside["Account Percent Change"] = 0
for row in range(len(sidebyside)):
    if sidebyside[0][row] == sidebyside[8][row]:
        if sidebyside[1][row] == sidebyside[9][row] + 1:
            if sidebyside[11][row] != 0:
                sidebyside["Account Percent Change"][row] = (sidebyside[3][row] - sidebyside[11][row])/sidebyside[11][row]

sidebyside.head()
```
